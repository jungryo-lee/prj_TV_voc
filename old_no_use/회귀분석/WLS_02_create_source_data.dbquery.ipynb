```sql
CREATE TABLE IF NOT EXISTS sandbox.z_jungryo_lee.buzz_model_meta AS (
    select concat_ws('_', array(country, year, trim(model))) as model_id
        , country
        , year
        , brand_name
        , model
        , d_type 
    from sandbox.z_jungryo_lee.agg_buzzmetrix_model_category_v1
    );
    
CREATE TABLE IF NOT EXISTS sandbox.z_jungryo_lee.buzz_sc_score_dataset AS (
  SELECT concat_ws('_', array(country, year, trim(model))) as model_id, category, total_count, avg_sc
  FROM sandbox.z_jungryo_lee.agg_buzzmetrix_model_category_v1
);

select * from sandbox.z_jungryo_lee.buzz_model_meta
```
