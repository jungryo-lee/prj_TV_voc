```sql
SELECT country, year, model, memo, sc_measurement, cate_2_depth
FROM kic_data_ods.buzzmetrix.buzzmetrix
WHERE 1=1 
  and country in ("United States", "United Kingdom")
  and year in (2024, 2025)
  and model in ('(LG) OLED B4', '(LG) OLED B5', '(LG) OLED C4', '(LG) OLED C5', '(LG) OLED G4', '(LG) OLED G5')
  and cate_2_depth in ('07-06. 리모컨 사용성', '07-08. 음성 인식/조작')
```
